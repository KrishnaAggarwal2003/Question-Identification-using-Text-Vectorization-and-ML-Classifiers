```python
from sklearn.preprocessing import LabelEncoder
#from data_preprocess import apply_preprocessing
import nltk
import pandas as pd
from vectorization_models import text_models
from classification_models import model_plots
import matplotlib.pyplot as plt
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data_frame = pd.read_csv('Project_data.csv')
data_frame['binary_label'] = LabelEncoder().fit_transform(data_frame['label'])
```

```python
x = data_frame['sentence']
y = data_frame['binary_label']
```

```python
models = ['TF-IDF Text Vectorization', 'One-word Embedding', 'LDA Text Classification', 'LSA Text Vectorization']
desired_model = ['tf-idf' , 'one-word','lda', 'lsa']
```

```python
for model_name in desired_model:
    x_train,x_test,y_train,y_test = text_models(x,y,model_name=[name for name in models if model_name.lower() in name.lower()][0])
    model_plots(x_train , x_test, y_train, y_test)
    plt.show()
```
